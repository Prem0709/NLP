## WhitespaceTokenizer
---
* `WhitespaceTokenizer` splits the input text **only** on whitespace characters like spaces, tabs, and newlines.
* Unlike `word_tokenize`, it **doesn't separate punctuation** from words. For example, `"systems."` will stay as `"systems."`, not `"systems"` and `"."`.

## Understanding N-grams in NLP: Bigrams, Trigrams & Beyond
---

The `nltk.util` module provides tools for generating **n-grams**, which are sequences of `n` consecutive items (usually words or characters) from a given text. You are specifically looking to use `bigrams`, `trigrams`, and `ngrams` from NLTK.
Here’s a brief overview of what these functions do:
* **Bigrams**: Pairs of consecutive words (2-grams).
* **Trigrams**: Triplets of consecutive words (3-grams).
* **Ngrams**: Sequences of `n` consecutive words, where `n` can be any number.


#### import bigram, trigram, ngram

````python
from nltk.util import bigrams, trigrams, ngrams
````

````python
string = 'hello the best and most beautifull thing in the world cannot be seen or even touched,they must be felt with heart'
quotes_tokens = nltk.word_tokenize(string)
quotes_tokens
````

````python
len(quotes_tokens)
````

### 🧠 **Explanation of N-grams**

---
#### Biagram 
---

1. **Bigrams**: Sequences of 2 consecutive tokens.
   * Example: For the text `"Artificial Intelligence is powerful"`, the bigrams are:
     ```
     ('Artificial', 'Intelligence'), 
     ('Intelligence', 'is'), 
     ('is', 'powerful')
     ```

````python
quotes_bigrams = list(nltk.bigrams(quotes_tokens))
quotes_bigrams
````

---
#### Trigrams
---
2. **Trigrams**: Sequences of 3 consecutive tokens.
   * Example: For the same text, the trigrams are:
     ```
     ('Artificial', 'Intelligence', 'is'), 
     ('Intelligence', 'is', 'powerful')
     ```

````python
quotes_trigrams = list(nltk.trigrams(quotes_tokens))
quotes_trigrams
````

````python
print(len(quotes_trigrams))
print(len(quotes_bigrams))
````

---
####  Ngrams
---
3. **Ngrams**: You can create n-grams for any number `n`. This is just a generalization of the bigram and trigram concepts.


````python
quotes_ngrams = list(nltk.ngrams(quotes_tokens))
quotes_ngrams
````

````python
quotes_ngrams = list(nltk.ngrams(quotes_tokens, 5))
quotes_ngrams
````

## Stemming
---
**Stemming** is a **text preprocessing technique** in **Natural Language Processing (NLP)** where words are **reduced to their root or base form**.

* It removes suffixes (like *ing*, *ed*, *s*) to get to the stem of the word.
* The stemmed word **might not always be a valid word** in English.

### types of Stemming:
1. Porter Stemmer
2. Lancaster Stemmer (Paice-Husk)
3. Snowball Stemmer (a.k.a. Porter2)
4. Regex-based Stemmer (Custom)
5. Lemmatizer (Alternative, not exactly stemming)

#### 1. Porter Stemmer
- Uses a set of rules to remove common English suffixes

    ✅ Example:
    `"running" → "run"`
    `"relational" → "relat"`

````python
from nltk.stem import PorterStemmer
pts = PorterStemmer()
print(pts.stem("premkumar"))
print(pts.stem("happiness"))
print(pts.stem("jumps"))
print(pts.stem("running"))
````

````python
words_to_stem = ["running", "jumps", "easily", "fishing", "happiness", "studies", "agreed", "faster", "crying", "relational"]
for words in words_to_stem:
    print(words + ' : ' + pts.stem(words))
````

#### 2. Lancaster Stemmer (Paice-Husk)

* 📌 **More aggressive** than Porter
* Can result in over-stemming (removing too much)
* Fast but less accurate

    ✅ Example:
    `"running" → "run"`
    `"maximum" → "maxim"`
    `"studies" → "stud"`

````python
from nltk.stem import LancasterStemmer
lct = LancasterStemmer()

for word in words_to_stem:
    print(word + " : " + lct.stem(word))
````

#### 3. Snowball Stemmer (a.k.a. Porter2)

* 📌 Improved and more international version of the Porter stemmer
* Supports multiple languages (English, French, German, etc.)
* **More consistent** and **flexible**

    ✅ Example:
    `"fishing" → "fish"`
    `"fished" → "fish"`

````python
from nltk.stem import SnowballStemmer
sbs = SnowballStemmer("english", "french")

for word in words_to_stem:
    print(word + " : " + sbs.stem(word))
````

#### 4. Regex-based Stemmer (Custom)**

* Uses regular expressions to manually strip suffixes
* Useful for domain-specific rules
* Not very flexible or general-purpose

````python
from nltk.stem import RegexpStemmer
rgx = RegexpStemmer('ing$|ed$|s$', min=4)

for word in words_to_stem:
    print(word + " : " + rgx.stem(word))
````

#### 5. Lemmatizer (Alternative, not exactly stemming)**

* Not a stemmer per se, but often mentioned alongside
* Uses **vocabulary + grammar rules** to return valid dictionary words
* **More accurate**, but slower

    ✅ Example:
    `"better" → "good"` (using WordNetLemmatizer)

````python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
words_to_lemmatize = ["better", "running", "fishing", "happiness", "studies"]

for word in words_to_lemmatize:
    print(word + " : " + lemmatizer.lemmatize(word, pos='a'))
````

## regular expressions module
---
The **`re` module** in Python provides full support for **regular expressions (regex)**, which are powerful tools for **searching**, **matching**, and **manipulating** text using **patterns**.


````python
import re

# Compile the pattern to remove punctuation and digits
punctuation = re.compile(r'[-.?!,:;()|0-9]')

# Example list of words (tokens)
ai_tokens = ['Artificial', 'Intelligence', 'refers', 'to', 'the', 'intelligence', 'of', 'machines', '.', 'This', 'is', 'AI', 'version2.0', '!', 'fastest-growing']

# Create a new list to store cleaned words
ai_cleaned = []

# Loop through each word and remove punctuation/digits
for word in ai_tokens:
    cleaned = punctuation.sub('', word)  # Replace matched characters with empty string
    if cleaned:  # Only add non-empty results
        ai_cleaned.append(cleaned)

# Print results
print("Original tokens:")
print(ai_tokens)
print("\nCleaned tokens:")
print(ai_cleaned)
print("\nLength of cleaned tokens:", len(ai_cleaned))
````

## Chunk
---
### 🧩 What is **Chunking** in NLP?
**Chunking**, also known as **shallow parsing**, is the process of extracting **phrases** from text by grouping **related tokens** based on their Part-of-Speech (POS) tags.

````python
from nltk import ne_chunk
NE_sent = "Artificial Intelligence is changing the world rapidly."

NE_tokens = word_tokenize(NE_sent)
NE_tokens
````

````python
NE_tags = nltk.pos_tag(NE_tokens)
NE_tags
````

````python
NE_NER = ne_chunk(NE_tags)
print(NE_NER)
````
