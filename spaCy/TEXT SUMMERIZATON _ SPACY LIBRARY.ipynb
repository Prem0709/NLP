```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
stopword = list(STOP_WORDS)
stopword
```

```python
len(stopword)
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
text
```

```python
doc= nlp(text)
doc
```

```python
# lets get the tokens from text
tokens = [token.text for token in doc]
print(tokens)
```

```python
len(tokens)
```

```python
print(doc)
```

```python
word_frequencies = {}  # Initialize an empty dictionary to store word counts

for word in doc:  # Iterate through each token (word) in the processed document
    if word.text.lower() not in stopword:  # Ignore stopwords (e.g., "the", "is")
        if word.text.lower() not in punctuation:  # Ignore punctuation (e.g., ".", ",")
            if word.text not in word_frequencies.keys():  # If the word is not yet in the dictionary
                word_frequencies[word.text] = 1  # Add it with an initial count of 1
            else:
                word_frequencies[word.text] += 1  # If already present, increment its count
```

```python
word_frequencies
#print(word_frequencies)
```

```python
len(word_frequencies)
```

```python
max_frequency = max(word_frequencies.values())
max_frequency
```

```python
#to get normalized/weighted frequencies you should devide all frequencies with 11
for word in word_frequencies.keys():
    word_frequencies[word] = word_frequencies[word]/max_frequency
```

```python
word_frequencies
```

```python
sentence_tokens = [sent for sent in doc.sents]  # Create a list of sentences from the document using spaCy's sentence segmentation
sentence_tokens  # Display the list of sentence tokens
```

```python
len(sentence_tokens)
```

```python
# We are going to calculate the score of each sentence based on the frequency of significant words
sentence_scores = {}  # Initialize an empty dictionary to store sentence scores

for sent in sentence_tokens:  # Iterate through each sentence in the list of sentence tokens
    for word in sent:  # Iterate through each word in the sentence
        if word.text.lower() in word_frequencies.keys():  # Check if the word is in our frequency dictionary (i.e., it's important)
            if sent not in sentence_scores.keys():  # If the sentence is not yet scored
                sentence_scores[sent] = word_frequencies[word.text.lower()]  # Start its score with the word's frequency
            else:
                sentence_scores[sent] += word_frequencies[word.text.lower()]  # Otherwise, add the word's frequency to the existing score
```

```python
sentence_scores
```

```python
from heapq import nlargest  # Import nlargest to get top N items from a dictionary

select_length = int(len(sentence_tokens) * 0.4)  # Select 40% of the total number of sentences
select_length  # This holds the number of top sentences you want to include in the summary
```

```python
# We have to select the top 'select_length' (e.g., 6) sentences with the highest scores from 'sentence_scores'
summary = nlargest(select_length, sentence_scores, key=sentence_scores.get)
```

```python
summary
```

```python
final_summary = [word.text for word in summary]
final_summary
```
